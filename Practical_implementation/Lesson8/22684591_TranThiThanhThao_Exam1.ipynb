```python
class Node:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None

    def __str__(self):
        print(str(self.data))

    def insert_node(self,value):
        if self.data == value:
            return False
        elif value < self.data:
            if self.left:
                return self.left.insert_node(value)
            else:
                self.left = Node(value)
                return True
        else:
            if self.right:
                return self.right.insert_node(value)
            else:
                self.right = Node(value)
                return True
            
    def min_value_node(self):
        current = self # bắt đầu từ node hiện tại
        while current.left is not None:
            current = current.left
        return current
    
    def max_value_node(self):
        current = self # bắt đầu từ node hiện tại
        while current.right is not None:
            current = current.right
        return current
    
    def delete_node(self,data):
        if data < self.data and self.left: #tìm dữ liệu data
            self.left = self.left.delete_node(data) #tìm dữ liệu data
        elif data > self.data and self.right: #tìm dữ liệu data
            self.right = self.right.delete_node(data) #tìm dữ liệu data
        else:
            #case 1: nó là nút lá (k có con)
            if self.left is None and self.right is None: #nếu 2 nhánh đều rỗng thì trả về Node. Tôi thắc mắc Data nhập vào đâu rồi
                return None
            #case 2: nút lá có 1 con:
            if self.left is None: #nếu left rỗng thì trả về right để thay thể giá trị Node(data) nhập vào.
                return self.right
            if self.right is None: #nếu right rỗng thì trả về left để thay thể giá trị Node(data) nhập vào.
                return self.left 
            #case 3: nút lá có 2 con, tìm giá trị nhỏ nhất bên phải
            temp = self.right # cho biến tạm là self.right
            while temp.left: #lặp giá trị nhỏ nhất bên phải
                temp = temp.left 
            self.data = temp.data #chuyển giá trị data là biến giá trị left đang nhỏ nhất
            #lúc này giá trị được gán vẫn còn tồn tại
            self.right = self.right.delete_node(temp.data) #đệ quy để xóa giá trị nhỏ nhất bên trái của bên phải 
        return self #trả về cây mới
    
    def update_node(self,data_old, data_new):
        #Tìm data ở vị trí nào
        if self.find(data_old):
            self.delete_node(data_old)
            self.insert_node(data_new)
            return True
        else:
            return False
    def find(self, value):
        if value == self.data: # nếu value tìm bằng dữ liệu của Node thì return tìm thấy
            return "Da tim thay"
        elif value < self.data: # value nhỏ hơn data của node 
            if self.left: #rẽ sang node trái để tìm
                return self.left.find(value) # Đệ quy để tìm cho tới khi thấy value (tức là có xét if value == self.data)
            else:
                return False # trả false nếu k tìm thấy
        elif value > self.data: #value lớn hơn data của node
            if self.right: #rẽ sang phải để tìm
                return self.right.find(value) #đệ quy để tìm cho tới khi thấy value (tức là có xét if value == self.data)
            else:
                return False # trả false nếu k tìm thấy
            
    def countLeaf(self):
        #khai báo đếm = 0
        dem = 0
        if self.left is None and self.right is None:
            return 1
        if self.left:
            dem += self.left.countLeaf()
        if self.right:
            dem += self.right.countLeaf()
        return dem
    
    def hightTree(self):
        hight_left = 0 #khai báo biến
        hight_right = 0
        if self.left:
            hight_left = self.left.hightTree() # duyệt left để chạy tới độ sâu nhất
        if self.right:
            hight_right = self.right.hightTree() # duyệt right để chạy tới độ sâu nhất
        return max(hight_left,hight_right) + 1 #dùng max để lấy giá trị cao nhất + 1 (root)
    
    def preNode(self): #Node left right
        if self.data:
            print(self.data, end = '  ')
        if self.left:
            self.left.preNode()
        if self.right:
            self.right.preNode()

    def inNode(self): # left Node right
        if self.left:
            self.left.inNode()
        if self.data:
            print(self.data, end = '---')
        if self.right:
            self.right.inNode()

    def afterNode(self): # left right Node
        if self.left:
            self.left.afterNode()
        if self.right:
            self.right.afterNode()
        if self.data:
            print(self.data, end = '  ')
```
