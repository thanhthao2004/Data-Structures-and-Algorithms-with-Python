Các lớp Nhân viên:
• Lớp Nhân viên (Employee):
o Lưu trữ thông tin cơ bản (id, tên, tuổi, email, lương cơ bản) 
o Cung cấp hàm tạo và phương thức biểu diễn chuỗi phù hợp
• Lớp Lập trình viên (Developer) (kế thừa từ Employee):
o Thêm thuộc tính: ngôn ngữ lập trình chính, số năm kinh nghiệm, cấp độ
(junior/senior)
o Ghi đè phương thức biểu diễn chuỗi
• Lớp Quản lý Dự án (ProjectManager) (kế thừa từ Employee):
o Thêm thuộc tính: số dự án đã hoàn thành, lĩnh vực chuyên môn, chứng chỉ 
o Ghi đè phương thức biểu diễn chuỗi
• Lớp Kiểm thử viên (Tester) (kế thừa từ Employee):
o Thêm thuộc tính: loại kiểm thử (manual/automation), công cụ sử dụng 
o Ghi đè phương thức biểu diễn chuỗi
Lớp Stack:
• Triển khai cấu trúc dữ liệu Stack với các phương thức sau: 
o push: Thêm một mục vào stack
o pop: Xóa và trả về mục trên cùng
o is_empty: Kiểm tra xem stack có trống không
o show: Hiển thị nội dung stack
Lớp Quản lý Nhân viên (EmployeeManagement):
• Sử dụng lớp Stack để quản lý nhân viên
• Triển khai các chức năng sau:
Chức năng yêu cầu:
• Thêm các loại nhân viên khác nhau vào hệ thống
• Xóa nhân viên
• Hiển thị danh sách nhân viên
• Tìm kiếm nhân viên theo ID
• Cập nhật thông tin nhân viên
• Tìm nhân viên có kinh nghiệm cao nhất
• Tìm quản lý dự án có nhiều dự án thành công nhất
• Sắp xếp nhân viên theo mức lương sử dụng Bubble Sort
• Triển khai chức năng hoàn tác (undo) các thao tác quản lý nhân viên bằng Stack
• Lưu dữ liệu nhân viên vào file
• Đọc dữ liệu nhân viên từ file

```python
class Employee:
    def __init__(self, id, name, age, email, salary):
        self.id = id
        self.name = name
        self.age = age
        self.email = email
        self.salary = salary
    def __str__(self):
        return f"id: {self.id}, tên: {self.name}, tuổi: {self.age}, email: {self.email}, lương cơ bản: {self.salary}"
    def tinhluong(self):
        return self.salary +40000
class Developer(Employee):
    def __init__(self,id, name, age, email, salary, main_language, ex_year, level):
        super().__init__(id, name, age, email, salary)
        self.main_language = main_language
        self.ex_year = ex_year
        self.level = level
    def __str__(self):
        return super().__str__() + f", ngôn ngữ lập trình chính: {self.main_language}, số năm kinh nghiệm: {self.ex_year}, cấp độ: {self.level}"
    def tinhluongdev(self):
        return super().tinhluong() + 50000
class ProjectManager(Employee):
    def __init__(self,id, name, age, email, salary, nums_project, major, certificate):
        super().__init__(id, name, age, email, salary)
        self.nums_project = nums_project
        self.major = major
        self.certificate = certificate
    def __str__(self):
        return super().__str__() + f", số dự án đã hoàn thành: {self.nums_project}, lĩnh vực chuyên môn: {self.major}, chứng chỉ: {self.certificate}"
class Tester(Employee):
    def __init__(self,id, name, age, email, salary, type_test, tool_use):
        super().__init__(id, name, age, email, salary)
        self.type_test = type_test
        self.tool_use = tool_use
    def __str__(self):
        return super().__str__() + f", loại kiểm thử: {self.type_test}, công cụ sử dụng: {self.tool_use}"
```

```python
class Stack:
    def __init__(self):
        self.stack = []
    def is_empty(self):
        return len(self.stack) == 0
    def push(self,item):
        return self.stack.insert(0,item)
    def pop(self):
        if self.is_empty():
            return "Không có gì"
        return self.stack.pop(0)
    def show(self):
        if self.is_empty():
            return "Không có gì"
        return str(self.stack)
```

```python
import csv

class EmployeeManagement:
    def __init__(self):
        self.ds = Stack()  # Stack chứa danh sách nhân viên
        self.history = Stack()  # Stack để lưu trạng thái trước mỗi thao tác (Undo)

    def add_employee(self):
        id = input("Nhập ID nhân viên: ")
        name = input("Nhập tên nhân viên: ")
        age = input("Nhập tuổi nhân viên: ")
        email = input("Nhập email nhân viên: ")
        salary = input("Nhập lương của nhân viên: ")
        return id, name, age, email, salary

    def add_type_employee(self):
        while True:
            print("""
                    1. Thêm nhân viên Lập trình viên
                    2. Thêm nhân viên Quản lý Dự án
                    3. Thêm nhân viên Kiểm thử viên
                    0. Thoát
                """)
            choose = int(input("Chọn loại nhân viên: "))
            if choose == 0:
                break
            else:
                n = int(input("Nhập số lượng nhân viên bạn muốn thêm: "))
                for _ in range(n):
                    id, name, age, email, salary = self.add_employee()
                    if choose == 1:
                        main_language = input("Nhập ngôn ngữ lập trình: ")
                        ex_year = int(input("Nhập số năm kinh nghiệm: "))
                        level = input("Nhập cấp độ junior/senior: ")
                        self.ds.push(Developer(id, name, age, email, salary, main_language, ex_year, level))
                    elif choose == 2:
                        nums_project = input("Nhập số dự án đã hoàn thành: ")
                        major = input("Nhập lĩnh vực chuyên môn: ")
                        certificate = input("Nhập chứng chỉ: ")
                        self.ds.push(ProjectManager(id, name, age, email, salary, nums_project, major, certificate))
                    elif choose == 3:
                        type_test = input("Nhập loại kiểm thử (manual/automation): ")
                        tool_use = input("Nhập công cụ sử dụng: ")
                        self.ds.push(Tester(id, name, age, email, salary, type_test, tool_use))
                    else:
                        print("Lựa chọn không hợp lệ!")

    def delete_employee(self):
        id_emp = input("Nhập ID nhân viên muốn xóa: ")
        temp = Stack()
        found = False

        self.history.push(self.ds.show())  # Lưu trạng thái trước khi xóa

        while not self.ds.is_empty():
            emp = self.ds.pop()
            if emp.id == id_emp:
                found = True
                print(f"Đã xóa nhân viên {id_emp}")
                break
            temp.push(emp)

        while not temp.is_empty():
            self.ds.push(temp.pop())

        if not found:
            print("Không tìm thấy nhân viên!")

    def display_listEmp(self):
        for emp in reversed(self.ds.show()):  # Duyệt theo thứ tự từ dưới lên
            print(emp)

    def find_employee(self, id_emp):
        temp = Stack()
        found = None

        while not self.ds.is_empty():
            emp = self.ds.pop()
            if emp.id == id_emp:
                found = emp
            temp.push(emp)

        while not temp.is_empty():
            self.ds.push(temp.pop())

        return found

    def update_employee(self):
        id_emp = input("Nhập ID nhân viên cần cập nhật: ")
        emp = self.find_employee(id_emp)

        if emp:
            print("Thông tin nhân viên:", emp)
            attr = input("Nhập thuộc tính cần cập nhật: ")
            if hasattr(emp, attr):
                new_value = input(f"Nhập giá trị mới cho {attr}: ")
                self.history.push(self.ds.show())  # Lưu trạng thái trước khi cập nhật
                setattr(emp, attr, new_value)
                print(f"Đã cập nhật {attr} thành {new_value}")
            else:
                print("Thuộc tính không hợp lệ!")
        else:
            print("Không tìm thấy nhân viên!")

    def find_most_experienced_developer(self):
        developers = [emp for emp in self.ds.show() if isinstance(emp, Developer)]
        if developers:
            most_experienced = max(developers, key=lambda dev: dev.ex_year)
            print("Lập trình viên có nhiều kinh nghiệm nhất:", most_experienced)
        else:
            print("Không có lập trình viên!")

    def find_most_successful_manager(self):
        managers = [emp for emp in self.ds.show() if isinstance(emp, ProjectManager)]
        if managers:
            most_successful = max(managers, key=lambda mgr: mgr.nums_project)
            print("Quản lý dự án thành công nhất:", most_successful)
        else:
            print("Không có quản lý dự án!")

    def sort_employees_by_salary(self):
        temp = self.ds.show()[:]
        n = len(temp)

        for i in range(n - 1):
            for j in range(n - 1 - i):
                if float(temp[j].salary) > float(temp[j + 1].salary):
                    temp[j], temp[j + 1] = temp[j + 1], temp[j]

        self.ds = Stack()
        for emp in temp:
            self.ds.push(emp)

        print("Đã sắp xếp nhân viên theo lương!")

    def undo(self):
        if not self.history.is_empty():
            self.ds = Stack()
            for emp in reversed(self.history.pop()):  # Phục hồi trạng thái trước đó
                self.ds.push(emp)
            print("Hoàn tác thành công!")
        else:
            print("Không có thao tác nào để hoàn tác!")

    def save_data_csv(self, filename):
        with open(filename, "w", newline="", encoding="utf-8") as f:
            writer = csv.writer(f)
            writer.writerow(["id", "name", "age", "email", "salary", "role"])
            for emp in self.ds.show():
                writer.writerow([emp.id, emp.name, emp.age, emp.email, emp.salary, emp.__class__.__name__])
        print(f"Dữ liệu đã được lưu vào {filename}!")

    def load_data_csv(self, filename):
        try:
            with open(filename, "r", encoding="utf-8") as f:
                reader = csv.DictReader(f)
                self.ds = Stack()
                for row in reader:
                    role = row["role"]
                    if role == "Developer":
                        emp = Developer(**row)
                    elif role == "ProjectManager":
                        emp = ProjectManager(**row)
                    elif role == "Tester":
                        emp = Tester(**row)
                    else:
                        continue
                    self.ds.push(emp)
            print(f"Dữ liệu đã được tải từ {filename}!")
        except FileNotFoundError:
            print("Không tìm thấy file!")
```

```python
ql = EmployeeManagement()
```

```python
while True:
    print("""   
                1. Thêm các loại nhân viên khác nhau vào hệ thống
                2. Xóa nhân viên
                3. Hiển thị danh sách nhân viên
                4. Tìm kiếm nhân viên theo ID
                5. Cập nhật thông tin nhân viên
                6. Tìm nhân viên có kinh nghiệm cao nhất
                7. Tìm quản lý dự án có nhiều dự án thành công nhất
                8. Sắp xếp nhân viên theo mức lương sử dụng Bubble Sort
                9. Triển khai chức năng hoàn tác (undo) các thao tác quản lý nhân viên bằng Stack
                10. Lưu dữ liệu nhân viên vào file
                11. Đọc dữ liệu nhân viên từ file
                0. Thoát chương trình
     """)
    
    choose = int(input("Nhập lựa chọn: "))
    
    if choose == 1:
        ql.add_type_employee()
    elif choose == 2:
        ql.delete_employee()
    elif choose == 3:
        ql.display_listEmp()
    elif choose == 4:
        id_emp = input("Nhập ID nhân viên cần tìm: ")
        emp = ql.find_employee(id_emp)
        if emp:
            print("Thông tin nhân viên:", emp)
        else:
            print("Không tìm thấy nhân viên!")
    elif choose == 5:
        ql.update_employee()
    elif choose == 6:
        ql.find_most_experienced_developer()
    elif choose == 7:
        ql.find_most_successful_manager()
    elif choose == 8:
        ql.sort_employees_by_salary()
    elif choose == 9:
        ql.undo()
    elif choose == 10:
        filename = input("Nhập tên file để lưu: ")
        ql.save_data_csv(filename)
    elif choose == 11:
        filename = input("Nhập tên file để đọc dữ liệu: ")
        ql.load_data_csv(filename)
    elif choose == 0:
        print("Thoát chương trình!")
        break
    else:
        print("Lựa chọn không hợp lệ! Vui lòng nhập lại.")
```
