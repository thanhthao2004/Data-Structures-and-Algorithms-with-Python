```python
class Sinhvien:
    def __init__(self, MaSV, Hoten, SDT, Diem):
        self.MaSV = MaSV
        self.Hoten = Hoten
        self.SDT = SDT
        self.Diem = Diem
    def In_Thong_Tin(self):
        print(f"Ma sinh vien: {self.MaSV}, Ho va ten: {self.Hoten}, SDT: {self.SDT}, Diem: {self.Diem}")
```

```python
def bubble_sort(arr, key):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j][key] > arr[j+1][key]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

def insertion_sort(arr, key):
    for i in range(1, len(arr)):
        key_item = arr[i]
        j = i - 1
        while j >= 0 and arr[j][key] > key_item[key]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key_item

def selection_sort(arr, key):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j][key] < arr[min_idx][key]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

def linear_search(students, MaSV):
    return next((sv for sv in students if sv["MaSV"] == MaSV), None)

def binary_search(students, SDT):
    sorted_students = sorted(students, key=lambda x: x["SDT"])
    left, right = 0, len(sorted_students) - 1
    while left <= right:
        mid = (left + right) // 2
        if sorted_students[mid]["SDT"] == SDT:
            return sorted_students[mid]
        elif sorted_students[mid]["SDT"] < SDT:
            left = mid + 1
        else:
            right = mid - 1
    return None

def save_to_file(students, filename="students.json"):
    with open(filename, "w") as f:
        json.dump(students, f, indent=4)
    print("Đã lưu vào file", filename)
```

```python
def menu():
    students = []
    while True:
        print("\n1. Nhập n sinh viên")
        print("2. In danh sách sinh viên")
        print("3. Tìm kiếm theo MSSV (Tuyến tính)")
        print("4. Tìm kiếm theo SDT (Nhị phân)")
        print("5. Tìm sinh viên có điểm cao nhất")
        print("6. Tìm sinh viên điểm cao nhất và thứ hai")
        print("7. Sắp xếp theo MSSV (Bubble Sort)")
        print("8. Sắp xếp theo Họ Tên (Insertion Sort)")
        print("9. Sắp xếp theo Điểm (Selection Sort)")
        print("10. Lưu vào file")
        print("11. Thoát")
        
        choice = input("Chọn chức năng: ")
        if choice == "1":
            n = int(input("Nhập số lượng sinh viên: "))
            for _ in range(n):
                students.append({
                    "MaSV": input("MSSV: "),
                    "HoTen": input("Họ Tên: "),
                    "SDT": input("Số ĐT: "),
                    "Diem": float(input("Điểm: "))
                })
        elif choice == "2":
            print(json.dumps(students, indent=4, ensure_ascii=False))
        elif choice == "3":
            print(linear_search(students, input("Nhập MSSV cần tìm: ")) or "Không tìm thấy sinh viên")
        elif choice == "4":
            print(binary_search(students, input("Nhập SDT cần tìm: ")) or "Không tìm thấy sinh viên")
        elif choice == "5":
            print(max(students, key=lambda x: x["Diem"], default="Danh sách trống"))
        elif choice == "6":
            sorted_students = sorted(students, key=lambda x: x["Diem"], reverse=True)
            print(sorted_students[:2] if sorted_students else "Danh sách trống")
        elif choice == "7":
            bubble_sort(students, "MaSV")
            print("Danh sách sau khi sắp xếp MSSV:", json.dumps(students, indent=4, ensure_ascii=False))
        elif choice == "8":
            insertion_sort(students, "HoTen")
            print("Danh sách sau khi sắp xếp Họ Tên:", json.dumps(students, indent=4, ensure_ascii=False))
        elif choice == "9":
            selection_sort(students, "Diem")
            print("Danh sách sau khi sắp xếp Điểm:", json.dumps(students, indent=4, ensure_ascii=False))
        elif choice == "10":
            save_to_file(students)
        elif choice == "11":
            break
        else:
            print("Chức năng chưa được triển khai")

menu()
```
