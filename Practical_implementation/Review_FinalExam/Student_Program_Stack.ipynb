CÀI ĐẶT VÀ XỬ LÝ CÂY NHỊ PHÂN TÌM KIẾM (BST) THÔNG TIN SINH VIÊN
Mỗi node trong cây là thông tin của một sinh viên, bao gồm:

MASV (Mã sinh viên) – Khóa chính

HOTEN (Họ tên)

MALOP (Mã lớp)

DIEMTB (Điểm trung bình)

Thực hiện các chức năng sau:

1. Nhập dữ liệu cho cây: 

Nhập thông tin các sinh viên.

Nếu MASV trùng, thông báo và yêu cầu nhập lại MASV khác.

Dữ liệu được lưu trong cây nhị phân tìm kiếm (BST), theo thứ tự MASV.

2. Duyệt và xuất dữ liệu của cây: LNR (Trái - Gốc - Phải)

3. Duyệt và xuất dữ liệu của cây: LRN (Trái - Phải - Gốc)

4. Duyệt và xuất dữ liệu của cây: NLR (Gốc - Trái - Phải)

5. In ra cây dưới dạng trực quan:

Xuất cây ra màn hình dưới dạng cây ngang hoặc cây phân cấp, để dễ nhìn cấu trúc cây.

6. Sắp xếp cây theo MASV:

(Thực chất là duyệt LNR để có danh sách tăng dần theo MASV).

7. Tìm và đếm số nút lá (node không có con).

8. Tính chiều cao của cây.

9. Đếm tổng số node của cây.

10. Chèn một node vào cây, sau mã số sinh viên nhập từ bàn phím.

11. Cập nhật thông tin một node dựa vào MASV nhập vào.

12. Xóa một node sinh viên theo MASV.

13. Tìm kiếm node theo MASV từ bàn phím.

14. Đếm số node có 2 con.

15. Đếm số node có 1 con.

16. Đếm số node có MASV chẵn và lẻ.

17. Tìm các node lá, đưa vào ngăn xếp (Stack).

18. Ghi file từ danh sách stack.

19. Đọc file, khôi phục lại stack và in ra màn hình.



```python
class Node:
    def __init__(self,masv, hoten, malop, diemtb):
        self.masv = masv
        self.hoten = hoten
        self.malop = malop
        self.diemtb = diemtb
        self.left = None
        self.right = None

    def insert_node(self, masv, hoten, malop, diemtb):
        if self.masv == masv:
            print("Mã sinh viên trùng")
            return
        if self.masv > masv:
            if self.left is None:
                self.left = Node(masv, hoten, malop, diemtb)
            else:
                self.left.insert_node(masv, hoten, malop, diemtb)
        elif self.masv < masv:
            if self.right is None:
                self.right = Node(masv, hoten, malop, diemtb)
            else:
                self.right.insert_node(masv, hoten, malop, diemtb)
    
    def LNR(self,level = 0):
        if self.left:
            self.left.LNR(level + 1)
        print("   " * level,f"MASV: {self.masv}, Họ tên: {self.hoten}, Mã lớp: {self.malop}, Điểm TB: {self.diemtb}")
        if self.right:
            self.right.LNR(level + 1)

    def LRN(self):
        if self.left:
            self.left.LRN()
        if self.right:
            self.right.LRN()
        print(f"MASV: {self.masv}, Họ tên: {self.hoten}, Mã lớp: {self.malop}, Điểm TB: {self.diemtb}")

    def NLR(self):
        print(f"MASV: {self.masv}, Họ tên: {self.hoten}, Mã lớp: {self.malop}, Điểm TB: {self.diemtb}")
        if self.left:
            self.left.NLR()
        if self.right:
            self.right.NLR()
    
    def SortbyMaSV(self, result = None):
        if result is None:
            result = []
        if self.left:
            self.left.SortbyMaSV(result)
        result.append({
            'masv' : self.masv,
            'hoten' : self.hoten,
            'malop' : self.malop,
            'diemtb' : self.diemtb
        })
        if self.right:
            self.right.SortbyMaSV(result)
        return result

    def count_leaf(self):
        dem = 0
        if self.left is None and self.right is None:
            print("Lá của cây là sinh viên có mã số: ", self.masv)
            return 1
        if self.left:
            dem = dem + self.left.count_leaf()
        if self.right:
            dem = dem + self.right.count_leaf()
        return dem
    
    def hight_tree(self):
        hight_left = 0
        hight_right = 0
        if self.left:
            hight_left +=  self.left.hight_tree()
        if self.right:
            hight_right += self.right.hight_tree()
        return max(hight_left,hight_right) + 1
    
    def count_node(self):
        dem = 0
        if self.left:
            dem = dem + self.left.count_node()
        if self.right:
            dem = dem + self.right.count_node()
        return dem + 1
    
    def count_node_2_children(self):
        dem = 0
        here = 0
        if self.left is not None and self.right is not None:
            here = 1
        if self.left:
            dem = dem + self.left.count_node_2_children()
        if self.right:
            dem = dem + self.right.count_node_2_children()
        return dem + here
    
    def count_node_1_children(self):
        dem = 0
        here = 0
        if (self.left is not None and self.right is None) or (self.right is not None and self.left is None):
            here = 1
        if self.left:
            dem = dem + self.left.count_node_1_children()
        if self.right:
            dem = dem + self.right.count_node_1_children() 
        return dem + here
    
    def find_node_le(self):
        dem = 0
        here = 0
        if(self.masv%2 != 0):
            here = 1
        if self.left:
            dem = dem + self.left.find_node_le()
        if self.right:
            dem = dem + self.right.find_node_le()
        return here + dem
    
    def find_node_chan(self):
        dem = 0
        here = 0
        if(self.masv%2 == 0):
            here = 1
        if self.left:
            dem = dem + self.left.find_node_chan()
        if self.right:
            dem = dem + self.right.find_node_chan()
        return here + dem
    
    def alter_insert_node(self):
        masv = int(input("Nhập mã sinh viên muốn chèn: "))
        new_masv = int(input("Nhập mã sinh viên muốn thêm: "))
        new_hoten = input("Nhập họ tên muốn thêm: ")
        new_malop = int(input("Nhập mã lớp muốn thêm: "))
        new_diemtb = float(input("Nhập điểm trung bình muốn thêm: "))
        if(self.masv==masv):
          return False
        if(self.masv > masv):
            if self.left:
                self.left.alter_insert_node(masv)
            else:
                self.left = Node(new_masv, new_hoten, new_malop, new_diemtb)
                return True
        if(self.masv < masv):
            if self.right:
                self.right.alter_insert_node(masv)
            else:
                self.right = Node(new_masv, new_hoten, new_malop, new_diemtb)
                return True
    
```
