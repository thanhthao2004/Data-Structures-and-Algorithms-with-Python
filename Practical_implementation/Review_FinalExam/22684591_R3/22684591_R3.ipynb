```python
class thietbi:
    def __init__(self, matb, tentb, giatb, soluong, namsx):
        self.matb = matb
        self.tentb = tentb
        self.giatb = giatb
        self.soluong = soluong
        self.namsx = namsx
    def __str__(self):
        return f"Mã thiết bị: {self.matb}, tên thiết bị: {self.tentb}, giá thiết bị: {self.giatb}, số lượng: {self.soluong}, năm sản xuất: {self.namsx}"
    
class thietbi_kimloai(thietbi):
    def __init__(self, matb, tentb, giatb, soluong, namsx, chatlieu):
        super().__init__(matb, tentb, giatb, soluong, namsx)
        self.chatlieu = chatlieu
    def __str__(self):
        return super().__str__() + f', chất liệu: {self.chatlieu}'

class thietbi_nhua(thietbi):
    def __init__(self, matb, tentb, giatb, soluong, namsx, loainhua):
        super().__init__(matb, tentb, giatb, soluong, namsx)
        self.loainhua = loainhua
    def __str__(self):
        return super().__str__() + f', loại nhựa: {self.loainhua}'
```

```python
class Stack:
    def __init__(self):
        self.ds = []
    def is_empty(self):
        return len(self.ds) == 0
    def push(self, item):
        self.ds.insert(0,item)
    def pop(self):
        if self.is_empty():
            return 'Rỗng'
        self.ds.pop(0)
    def __str__(self):
        for i in self.ds:
            print(i)
```

```python
class bst:
    def __init__(self, data = None):
        self.data = data
        self.left = None
        self.right = None
    
    def add_node(self, data):
        if self.data is None:
            self.data = data
        elif self.data.matb == data.matb:
            print("Trùng mã!!!!")
            return False
        elif self.data.matb < data.matb:
            if self.right is None:
                self.right = bst(data)
                print(f'Thêm sang bên phải của {self.data.matb}')
                return True
            else:
                return self.right.add_node(data)
        elif self.data.matb > data.matb:
            if self.left is None:
                self.left = bst(data)
                print(f'Thêm sang bên trái của {self.data.matb}')
                return True
            else:
                return self.left.add_node(data)
    
    def read_and_insert(self):
        with open('devices.txt','r') as file:
            for i in file:
                part = i.strip().split(',')
                matb = int(part[0].strip())
                tentb = part[1].strip()
                giatb = float(part[2].strip())
                soluong = int(part[3].strip())
                namsx = int(part[4].strip())
                tb = thietbi(matb, tentb, giatb, soluong, namsx)
                self.add_node(tb)
    
    def add_node_type(self):
        print("""
            0. Thoát
            1. Thiết bị kim loại
            2. Thiết bị nhựa
        """)
        chon = int(input("Chọn 1 kiểu thiết bị muốn thêm: "))
        if chon == 0:
            return
        n = int(input("Chọn so luong muon them: "))
        for i in range(n):
            matb = int(input("Nhập mã thiết bị: "))
            tentb = input("Nhập tên thiết bị: ")
            giatb = float(input("Nhập giá thiết bị: "))
            soluong = int(input("Nhập số lượng: "))
            namsx = int(input("Nhập năm sản xuất: "))
            if chon == 1:
                chatlieu = input("Nhập chất liệu: ")
                tbkimloai = thietbi_kimloai(matb, tentb, giatb, soluong, namsx, chatlieu)
                self.add_node(tbkimloai)
            elif chon == 2:
                loainhua = input("Nhập loai nhua: ")
                tbnhua = thietbi_nhua(matb, tentb, giatb, soluong, namsx, loainhua)
                self.add_node(tbnhua)
                
    def NLR(self):
        print(self.data)
        if self.left:
            self.left.NLR()
        if self.right:
            self.right.NLR()
    def LNR(self):
        if self.left:
            self.left.LNR()
        print(self.data)
        if self.right:
            self.right.LNR()
    def LRN(self):
        if self.left:
            self.left.LRN()
        if self.right:
            self.right.LRN()
        print(self.data)
    
    def count_node_1_child(self):
        dem = 0
        if (self.left and self.right is None) or (self.right and self.left is None):
            dem = 1
        if self.left:
            dem += self.left.count_node_1_child()
        if self.right:
            dem += self.right.count_node_1_child()
        return dem
    
    def thietbi_lon_hon_x(self,x, s = None):
        if s is None:
            s = Stack()
        if self.data and self.data.matb > x:
            s.push(self.data)
        if self.left:
            self.left.thietbi_lon_hon_x(x,s)
        if self.right:
            self.right.thietbi_lon_hon_x(x,s)  
        return s

    def thietbi_giacao(self, ds = None):
        if ds is None:
            ds = []
        if self.data:
            ds.append(self.data)
        if self.left:
            self.left.thietbi_giacao(ds)
        if self.right:
            self.right.thietbi_giacao(ds) 
        return sorted(ds,key = lambda tb: tb.giatb , reverse= True) #tăng dần
    
    def delete_node(self, matb):
        if self.left and self.data.matb > matb:
            self.left = self.left.delete_node(matb)
        elif self.right and self.data.matb < matb:
            self.right = self.right.delete_node(matb)
        else:
            if self.left is None and self.right is None:
                return None
            elif self.left is None:
                return self.right
            elif self.right is None:
                return self.left
            temp = self.right
            while temp.left:
                temp = temp.left
            self.data = temp.data
            self.right = self.right.delete_node(temp.data.matb)
        return self
    
    def find_node_2_child(self):
        if self.left and self.right:
            with open('NodesWithTwoChild.txt','a') as file:
                file.write(str(self.data) + '\n')
        if self.left:
            self.left.find_node_2_child()
        if self.right:
            self.right.find_node_2_child()

    def update_thietbi(self,matb):
        if self.left and self.data.matb > matb:
            self.left = self.left.update_thietbi(matb)
        elif self.right and self.data.matb < matb:
            self.right = self.right.update_thietbi(matb)
        else:
            soluongmoi = int(input("Nhập số lượng mới: "))
            giamoi = float(input("Nhập giá mới: "))
            self.data.giatb = giamoi
            self.data.soluong = soluongmoi

    def write_file_LNR(self,ds = None):
        if ds is None:
            ds = []
        if self.left:
            self.left.write_file_LNR(ds)
        ds.append(self.data)
        if self.right:
            self.right.write_file_LNR(ds)
        return ds
    
    def result_write_file_LNR(self):
        kq = self.write_file_LNR()
        with open('output.txt','w') as file:
            for i in kq:
                file.write(str(i)+'\n')
```

```python
ql = bst()
while True:
    print(
        """
        1.	Đọc file devices.txt đổ vào cây nhị phân tìm kiếm (1đ) 
            o	Mỗi dòng trong file có định dạng: Mã thiết bị, Tên thiết bị, Giá thiết bị, Số lượng, Năm sản xuất
            o	Tối thiểu file phải có 6 dòng
        2. Thêm node vào cây với 2 loại thiết bị: thiết bị kim loại, thiết bị nhựa
        3.	Duyệt và xuất dữ liệu của cây nhị phân tìm kiếm theo các thứ tự: 
            o	Thứ tự NLR (Pre-order traversal) (1đ)
            o	Thứ tự LNR (In-order traversal) (1đ)
            o	Thứ tự LRN (Post-order traversal) (1đ)
        4.	Đếm số nút có 1 con của cây (1đ)
        5.	Tìm các thiết bị có giá lớn hơn x và lưu vào Stack (1.5đ) 
            o	x là giá trị do người dùng nhập
        6.	Tìm k thiết bị có giá cao nhất (1.5đ) 
            o	k là số lượng do người dùng nhập
        7.	Xoá một node có giá trị là Mã thiết bị do người dùng nhập (1.5đ)
        8.	Tìm những nút có 2 con, in ra file NodesWithTwoChild.txt (1.5đ)
        9.	Cập nhật thông tin (số lượng, giá) của một thiết bị dựa vào mã thiết bị (1đ)
        10.	Ghi toàn bộ dữ liệu của cây (theo thứ tự LNR) ra file output.txt (1đ)
        11.	Thoát. 
        """
    )
    choose = int(input("Nhập lựa chọn: "))
    if choose == 11:
        break
    elif choose == 1:
        ql.read_and_insert()
    elif choose == 2:
        ql.add_node_type()
    elif choose == 3: 
        print("Thứ tự NLR: \n")
        ql.NLR()
        print("Thứ tự LNR: \n")
        ql.LNR()
        print("Thứ tự LRN: \n")
        ql.LRN()
    elif choose == 4:
        kq = ql.count_node_1_child()
        print("Số nút lá có 1 con: ", kq)

    elif choose == 5:
        x = int(input("Nhập mã thiết bị x: "))
        s = ql.thietbi_lon_hon_x(x)
        print("Các thiết bị có giá trị lớn hơn là: ")
        s.__str__()
    
    elif choose == 6:
        k = int(input("Nhập k: "))
        kq = ql.thietbi_giacao()
        for i in kq[:k]:
            print(i)
    
    elif choose == 7:
        matb = int(input("Nhập mã tb muốn xóa: "))
        ql.delete_node(matb)
        print("Xóa thành công")

    elif choose == 8:
        ql.find_node_2_child()
        print("Thành công!!")

    elif choose == 9:
        matb = int(input("Nhập mã thiết bị cần cập nhật số lượng và giá: "))
        ql.update_thietbi(matb)
        print("Thành công!!")

    elif choose == 10:
        ql.result_write_file_LNR()
        print("Thành công!!!")
```
