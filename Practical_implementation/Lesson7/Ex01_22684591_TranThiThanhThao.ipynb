```python
class Student:
    def __init__(self, id, name,major,gpa):
        self.id = id
        self.name = name
        self.major = major
        self.gpa = gpa
    def __str__(self):
        return f'ID: {self.id}, Name: {self.name}, Major: {self.major}, GPA: {self.gpa}'
```

```python
class  Linkedlist_Student:
    class Node(Student):
        def __init__(self, id, name,major,gpa):
            super().__init__(id, name,major,gpa)
            self.next = None
        def __str__(self):
            return super().__str__()
        
    def __init__(self):
        self.head = None

    # input student
    def info_Student(self):
        id = int(input("Nhập id sinh viên: "))
        name = input("Nhập name sinh viên: ")
        major = input("Nhập major: ")
        gpa = int(input("Nhập GPA: "))
        return id,name,major,gpa

    # Kiểm tra rỗng
    def is_empty(self):
        return self.head is None
    
    # Add a new student to the beginning of the list
    def first_node(self):
        id,name,major,gpa = self.info_Student()
        new_node = self.Node(id,name,major,gpa)
        if self.is_empty():
            self.head = new_node
        else:
            new_node.next = self.head
            self.head = new_node

    # Add a new student to the end of the lis
    def end_node(self):
        id,name,major,gpa = self.info_Student()
        new_node = self.Node(id,name,major,gpa)
        if self.is_empty():
            self.head = new_node
        else:
            c = self.head #gán biến tạm là node đầu
            while c.next: # lặp node khi giá trị không null
               c = c.next # gán c là giá trị trỏ tới phần cuối cùng của Linked list mà trò tới NULL
            c.next = new_node # Thoát vòng lặp, gán giá trị NULL là 1 node mới
    
    # Add a new student after a specific student (by ID)
    def after_node(self):
        id_node = int(input("Nhập ID sinh viên bạn muốn chèn một sinh viên mới đằng sau: "))
        c = self.head
        while c and c.id != id_node:
            c = c.next
        #thoát vòng lặp khi tìm thấy giá trị id nhập vào hợp lệ với danh sách lk
        if c:
            id,name,major,gpa = self.info_Student()
            new_node = self.Node(id,name,major,gpa)
            new_node.next = c.next #chuyển cục dữ liệu đằng sau c vào đằng sau new node
            c.next = new_node #chèn new node vào sau c
        else:
            print("Không tìm thấy sinh viên đã nhập")
    
    # Display all students in the system
    def print_all_student(self):
        arr = []
        c = self.head
        while c:
            arr.append(c)
            c = c.next
        print("->".join(map(str,arr)))
    
    # Find a student by ID 
    def findByID(self):
        id_node = int(input("Nhập ID sinh viên bạn muốn tìm: "))
        c = self.head
        while c and id_node != c.id:
            c = c.next
        if c:
            c.__str__()
        else:
            print("Sinh viên không tồn tại")

    # Update a student's information (name, major, GPA)
    def updateStudent(self):
        id_node = int(input("Nhập ID sinh viên bạn muốn cập nhật: "))
        c = self.head
        while c and c.id != id_node:
            c = c.next
        if c:
            name = input("Nhập tên thay đổi: ")
            major = input("Nhập major thay đổi: ")
            gpa = int(input("Nhập gpa thay đổi: "))
            c.name = name
            c.major = major
            c.gpa = gpa
            print("Cập nhật thông tin sinh viên thành công!")
        else:
            print("Sinh viên không tồn tại")

    # Remove a student from the beginning of the list
    def remove_firt_student(self):
        if self.is_empty():
            print("Danh sách rỗng không thể xóa ")
        else:
            c = self.head
            self.head = c.next
            print("Sinh viên đầu tiên đã được xóa khỏi danh sách.")
    
    # Remove a student from the end of the list 
    def remove_end_student(self):
        if self.is_empty():
            print("Danh sách rỗng không thể xóa ")
        else:
            c = self.head
            if c.next is None:
                self.head = None
                print("Sinh viên cuối đã được xóa khỏi danh sách.")
            else:
                while c.next and c.next.next :
                    c = c.next
                c.next = None
                print("Sinh viên cuối đã được xóa khỏi danh sách.")
    
    # Remove a specific student (by ID)
    def remove_by_id(self):
        id_node = int(input("Nhập ID sinh viên bạn muốn xóa: "))
        c = self.head
        prev = None
        while c and c.id != id_node:
            prev = c # prev lấy giá trị trước c
            c = c.next # Chạy c để tìm
        if prev is None: #tức là vòng while sẽ không chạy nên prev None
            #Node nằm đầu
            self.head = c.next #chuyển sang Node tiếp theo
        else: #prev không None
            #Node nằm giữa / cuối
            prev.next = c.next # Bây giờ prev.next đang là c của đoạn c = c.next
            # nối c.next để không bị mất các node sau đó
        print(f"Đã xóa sinh viên có ID {id_node}")

    # Sort students by GPA (ascending and descending)
    def sort_by_gpa(self, ascending=True):
        if self.is_empty() or self.head.next is None:
            print("Không đủ sinh viên để sắp xếp.")
            return
        # Bubble sort trên linked list
        swapped = True
        while swapped:
            swapped = False
            current = self.head
            while current.next:
                if (ascending and current.gpa > current.next.gpa) or (not ascending and current.gpa < current.next.gpa):
                    # Hoán đổi dữ liệu
                    current.id, current.next.id = current.next.id, current.id
                    current.name, current.next.name = current.next.name, current.name
                    current.major, current.next.major = current.next.major, current.major
                    current.gpa, current.next.gpa = current.next.gpa, current.gpa
                    swapped = True
                current = current.next
        print("Đã sắp xếp danh sách sinh viên theo GPA.")

    # Find students with the highest and lowest GPAs
    def find_max_min_gpa(self):
        if self.is_empty():
            print("Danh sách rỗng.")
            return
        max_student = min_student = self.head
        current = self.head.next
        while current:
            if current.gpa > max_student.gpa:
                max_student = current
            if current.gpa < min_student.gpa:
                min_student = current
            current = current.next
        print("Sinh viên có GPA cao nhất:")
        print(max_student)
        print("Sinh viên có GPA thấp nhất:")
        print(min_student)

    # Generate a report showing how many students are enrolled in each major
    def report_by_major(self):
        if self.is_empty():
            print("Danh sách rỗng.")
            return
        major_count = {}
        current = self.head
        while current:
            if current.major in major_count:
                major_count[current.major] += 1
            else:
                major_count[current.major] = 1
            current = current.next
        print("Báo cáo số lượng sinh viên theo chuyên ngành:")
        for major, count in major_count.items():
            print(f"{major}: {count} sinh viên")

    # Save all student data to a file
    def save_to_file(self, filename="students.txt"):
        with open(filename, "w", encoding="utf-8") as f:
            current = self.head
            while current:
                line = f"{current.id},{current.name},{current.major},{current.gpa}\n"
                f.write(line)
                current = current.next
        print(f"Dữ liệu đã được lưu vào {filename}")

    # Load student data from a file
    def load_from_file(self, filename="students.txt"):
        try:
            with open(filename, "r", encoding="utf-8") as f:
                self.head = None
                for line in f:
                    id, name, major, gpa = line.strip().split(",")
                    new_node = self.Node(int(id), name, major, int(gpa))
                    if self.is_empty():
                        self.head = new_node
                    else:
                        c = self.head
                        while c.next:
                            c = c.next
                        c.next = new_node
            print(f"Dữ liệu đã được tải từ {filename}")
        except FileNotFoundError:
            print("File không tồn tại.")

    # Get the total number of students in the list 
    def count_students(self):
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next
        print(f"Tổng số sinh viên trong danh sách: {count}")
```

```python
management_sv = Linkedlist_Student()
while(True):
    print("""
        1. Exit()
        2. Add a new student to the beginning of the list 
        3. Add a new student to the end of the list 
        4. Add a new student after a specific student (by ID) 
        5. Display all students in the system 
        6. Find a student by ID 
        7. Update a student's information (name, major, GPA) 
        8. Remove a student from the beginning of the list 
        9. Remove a student from the end of the list 
        10. Remove a specific student (by ID) 
        11. Sort students by GPA (ascending and descending) 
        12. Find students with the highest and lowest GPAs 
        13. Generate a report showing how many students are enrolled in each major 
        14. Save all student data to a file 
        15. Load student data from a file 
        16. Check if the list is empty 
        17. Get the total number of students in the list 
    """)
    choose = int(input("Nhập lựa chọn menu"))
    if choose == 2:
        n = int(input("Nhập số sinh viên muốn thêm: "))
        for i in range(n):
            management_sv.first_node()
    elif choose == 3:
        n = int(input("Nhập số sinh viên muốn thêm: "))
        for i in range(n):
            management_sv.end_node()
    elif choose == 4:
        n = int(input("Nhập số sinh viên muốn thêm: "))
        for i in range(n):
            management_sv.after_node()
    elif choose == 5:
        management_sv.print_all_student()
    elif choose == 6:
        management_sv.findByID()
    elif choose == 7:
        management_sv.updateStudent()
    elif choose == 8:
        management_sv.remove_firt_student()
    elif choose == 9:
        management_sv.remove_end_student()
    elif choose == 10:
        management_sv.remove_by_id()
    elif choose == 11:
        sort_type = input("Sắp xếp tăng (asc) hay giảm (desc)? ")
        management_sv.sort_by_gpa(ascending=(sort_type == "asc"))
    elif choose == 12:
        management_sv.find_max_min_gpa()
    elif choose == 13:
        management_sv.report_by_major()
    elif choose == 14:
        management_sv.save_to_file()
    elif choose == 15:
        management_sv.load_from_file()
    elif choose == 16:
        print("Danh sách rỗng." if management_sv.is_empty() else "Danh sách không rỗng.")
    elif choose == 17:
        management_sv.count_students()
    elif choose == 1:
        print("Đã thoát chương trình.")
        break
    else:
        print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")
```
