```python
class Sach:
    id_counter = 1

    def __init__(self, ten, tac_gia, nam_xuat_ban, gia, so_luong):
        self.id = f"B{Sach.id_counter:03d}"
        Sach.id_counter += 1
        self.ten = ten
        self.tac_gia = tac_gia
        self.nam_xuat_ban = nam_xuat_ban
        self.gia = gia
        self.so_luong = so_luong

    def __str__(self):
        return f"ID: {self.id}, Tên: {self.ten}, Tác giả: {self.tac_gia}, Năm XB: {self.nam_xuat_ban}, Giá: {self.gia}, Số lượng: {self.so_luong}"
```

```python
class QuanLySach:
    def __init__(self):
        self.danh_sach_sach = []

    def them_sach(self):
        ten = input("Nhập tên sách: ")
        tac_gia = input("Nhập tác giả: ")
        nam_xuat_ban = int(input("Nhập năm xuất bản: "))
        gia = float(input("Nhập giá sách: "))
        so_luong = int(input("Nhập số lượng sách: "))
        sach = Sach(ten, tac_gia, nam_xuat_ban, gia, so_luong)
        self.danh_sach_sach.append(sach)
        print("Thêm sách thành công!")

    def hien_thi_danh_sach_sach(self):
        for sach in self.danh_sach_sach:
            print(sach)

    def sap_xep_theo_ten(self):
        n = len(self.danh_sach_sach)
        for i in range(n - 1):
            for j in range(n - i - 1):
                if self.danh_sach_sach[j].ten > self.danh_sach_sach[j + 1].ten:
                    self.danh_sach_sach[j], self.danh_sach_sach[j + 1] = self.danh_sach_sach[j + 1], self.danh_sach_sach[j]

    def sap_xep_theo_nam_xuat_ban(self):
        n = len(self.danh_sach_sach)
        for i in range(1, n):
            key = self.danh_sach_sach[i]
            j = i - 1
            while j >= 0 and self.danh_sach_sach[j].nam_xuat_ban > key.nam_xuat_ban:
                self.danh_sach_sach[j + 1] = self.danh_sach_sach[j]
                j -= 1
            self.danh_sach_sach[j + 1] = key

    def sap_xep_theo_gia(self):
        n = len(self.danh_sach_sach)
        for i in range(n):
            min_index = i
            for j in range(i + 1, n):
                if self.danh_sach_sach[j].gia < self.danh_sach_sach[min_index].gia:
                    min_index = j
            self.danh_sach_sach[i], self.danh_sach_sach[min_index] = self.danh_sach_sach[min_index], self.danh_sach_sach[i]

    def tim_kiem_sach(self):
        tu_khoa = input("Nhập ID hoặc tên sách cần tìm: ")
        for sach in self.danh_sach_sach:
            if sach.id == tu_khoa or sach.ten.lower() == tu_khoa.lower():
                print("Tìm thấy sách:")
                print(sach)
                return
        print("Không tìm thấy sách!")

    def muon_tra_sach(self):
        id_sach = input("Nhập ID sách cần mượn/trả: ")
        for sach in self.danh_sach_sach:
            if sach.id == id_sach:
                lua_chon = input("Nhập 'm' để mượn sách hoặc 't' để trả sách: ")
                if lua_chon == 'm':
                    if sach.so_luong > 0:
                        sach.so_luong -= 1
                        print("Mượn sách thành công!")
                    else:
                        print("Sách đã hết!")
                elif lua_chon == 't':
                    sach.so_luong += 1
                    print("Trả sách thành công!")
                return
        print("Không tìm thấy sách!")
```

```python
def menu():
    qls = QuanLySach()
    while True:
        print("\nQUẢN LÝ THƯ VIỆN")
        print("1. Thêm sách mới")
        print("2. Hiển thị danh sách sách")
        print("3. Sắp xếp theo tên sách (Bubble Sort)")
        print("4. Sắp xếp theo năm xuất bản (Insertion Sort)")
        print("5. Sắp xếp theo giá sách (Selection Sort)")
        print("6. Tìm kiếm sách theo ID hoặc tên")
        print("7. Mượn/Trả sách")
        print("8. Thoát")
        
        choice = input("Chọn chức năng: ")
        if choice == "1":
            qls.them_sach()
        elif choice == "2":
            qls.hien_thi_danh_sach_sach()
        elif choice == "3":
            qls.sap_xep_theo_ten()
            print("Danh sách đã được sắp xếp theo tên sách.")
        elif choice == "4":
            qls.sap_xep_theo_nam_xuat_ban()
            print("Danh sách đã được sắp xếp theo năm xuất bản.")
        elif choice == "5":
            qls.sap_xep_theo_gia()
            print("Danh sách đã được sắp xếp theo giá sách.")
        elif choice == "6":
            qls.tim_kiem_sach()
        elif choice == "7":
            qls.muon_tra_sach()
        elif choice == "8":
            print("Thoát chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ, vui lòng nhập lại.")

menu()
```
